```python
import json
import pandas as pd
```

```python
file_path = "data/laureates-1000.json"

with open(file_path, 'r') as file:
    data = json.load(file)

laureates = data['laureates']
```

```python
# Extract relevant information for each laureate
laureates_info = []

for laureate in data['laureates']:
    birth_place = laureate.get('birth', {}).get('place', {})
    birth_city = birth_place.get('city', {}).get('en')
    birth_country = birth_place.get('country', {}).get('en')
    birth_continent = birth_place.get('continent', {}).get('en')
    birth_lat = birth_place.get('cityNow', {}).get('latitude')
    birth_lon = birth_place.get('cityNow', {}).get('longitude')

    for prize in laureate.get('nobelPrizes', []):
        category = prize.get('category', {}).get('en')
        if category in ['Literature', 'Peace']:  # Exclude 'Literature' and 'Peace'
            continue
        for affiliation in prize.get('affiliations', []):
            affil_city = affiliation.get('city', {}).get('en')
            affil_country = affiliation.get('country', {}).get('en')
            affil_continent = affiliation.get('continent', {}).get('en')
            affil_lat = affiliation.get('cityNow', {}).get('latitude')
            affil_lon = affiliation.get('cityNow', {}).get('longitude')

            laureates_info.append({
                'name': laureate['knownName']['en'],
                'birth_city': birth_city,
                'birth_country': birth_country,
                'birth_continent': birth_continent,
                'birth_lat': birth_lat,
                'birth_lon': birth_lon,
                'affil_city': affil_city,
                'affil_country': affil_country,
                'affil_continent': affil_continent,
                'affil_lat': affil_lat,
                'affil_lon': affil_lon,
                'category': category,
            })
```

```python
len(laureates_info)
```

```python
# Convert the extracted information to a DataFrame
df = pd.DataFrame(laureates_info)
df.head()
```

```python
# Extract all unique category names
categories = set()
for category in df['category']:
    categories.add(category)

print(categories)
```

```python
from geopy.distance import geodesic
```

```python
# Calculate the distance between birth place and affiliation place
def calculate_distance(row):
    if pd.notnull(row['birth_lat']) and pd.notnull(row['birth_lon']) and pd.notnull(row['affil_lat']) and pd.notnull(row['affil_lon']):
        birth_coords = (row['birth_lat'], row['birth_lon'])
        affil_coords = (row['affil_lat'], row['affil_lon'])
        return geodesic(birth_coords, affil_coords).kilometers
    else:
        return None
```

```python
# Apply the distance calculation to the DataFrame
df['distance_km'] = df.apply(calculate_distance, axis=1)
```

```python
df.head()
```

```python
# Sort the DataFrame by distance in descending order to manually inspect the highest distances
df_sorted = df.sort_values(by='distance_km', ascending=False)
```
